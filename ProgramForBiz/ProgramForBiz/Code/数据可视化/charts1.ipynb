```python
from pyecharts import options as opts
from pyecharts.charts import Bar, Grid, Line, Liquid, Page, Pie, Sankey
from pyecharts.commons.utils import JsCode
from pyecharts.components import Table
from pyecharts.globals import ThemeType


pie_Ethnicity = (
    Pie({"theme": ThemeType.ROMA})
    .add(
        "",
        [list(z) for z in zip(ethnicity.index.tolist(),ethnicity.values.tolist())],
        radius=["25%", "65%"],
        center=["50%", "50%"],
        rosetype="radius",
        label_opts=opts.LabelOpts(is_show=False),
    )
    .set_global_opts(title_opts=opts.TitleOpts(title="Ethnicity"))
)

pie_Gender = (
    Pie({"theme": ThemeType.ROMA})
    .add(
        "",
        [list(z) for z in zip(
            ['female','male'],gender.values.tolist())],
        radius=["25%", "65%"],
        center=["50%", "50%"],
        rosetype="radius",
        label_opts=opts.LabelOpts(is_show=False),
    )
    .set_global_opts(title_opts=opts.TitleOpts(title="Gender"))
)

pie_Country = (
    Pie({"theme": ThemeType.ROMA})
    .add(
        "",
        [list(z) for z in zip(Country.index.tolist(),Country.values.tolist())],
        radius=["25%", "65%"],
        center=["50%", "50%"],
        rosetype="radius",
        label_opts=opts.LabelOpts(is_show=False),
    )
    .set_global_opts(title_opts=opts.TitleOpts(title="Country"))
)
```

```python
austim_jundice = df_aurism.groupby(['austim','jundice']).result.count()
jundice_Class = df_aurism.groupby(['jundice','Class.ASD']).result.count()
df_austim_jundice = pd.DataFrame(austim_jundice).reset_index()
df_jundice_Class = pd.DataFrame(jundice_Class).reset_index()
df_austim_jundice['jundice'] = df_austim_jundice['jundice'].map({'no':'jundice_no','yes':'jundice_yes'})
df_austim_jundice['austim'] = df_austim_jundice['austim'].map({'no':'austim_no','yes':'austim_yes'})
df_jundice_Class['jundice'] = df_jundice_Class['jundice'].map({'no':'jundice_no','yes':'jundice_yes'})
df_jundice_Class['Class.ASD'] = df_jundice_Class['Class.ASD'].map({'NO':'Class.ASD_no','YES':'Class.ASD_yes'})
df_austim_jundice_Class = pd.merge(df_austim_jundice,df_jundice_Class,on = 'jundice')
df_austim_jundice_Class
```

```python
nodes = []
for i in df_austim_jundice_Class["jundice"].unique():
    dic = {}
    dic["name"] = i
    nodes.append(dic)
    
for i in df_austim_jundice_Class["austim"].unique():
    dic = {}
    dic["name"] = i
    nodes.append(dic) 

for i in df_austim_jundice_Class["Class.ASD"].unique():
    dic = {}
    dic["name"] = i
    nodes.append(dic) 

links = []
for i in df_austim_jundice_Class.values:
    dic = {}
    dic['source'] = i[0]
    dic['target'] = i[1]
    dic['value'] = i[2]
    links.append(dic)
for i in df_austim_jundice_Class.values:
    dic = {}
    dic['source'] = i[1]
    dic['target'] = i[3]
    dic['value'] = i[4]
    links.append(dic)
links
```

```python
austim_jundice_Class_Sankey = (
    Sankey({"theme": ThemeType.ROMA})
    .add(
        '',
        nodes,
        links,
        linestyle_opt=opts.LineStyleOpts(opacity=0.2, curve=0.5, color="target"),
        label_opts=opts.LabelOpts(position="right"),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Austim_Jundice_Class.ASD"),
        tooltip_opts=opts.TooltipOpts(trigger="item", trigger_on="mousemove"),
    )
)
```

```python
austim_jundice_Class_Sankey.render_notebook()
```

```python
df_age = pd.read_csv('/Users/wangxiaoningsi/Desktop/Term1/6104program/final report/aurism_2.csv')
female_age_count = df_age[df_age['gender'] == 'f']
male_age_count = df_age[df_age['gender'] == 'm']
female_age_count = female_age_count.groupby(['age']).result.count()
male_age_count = male_age_count.groupby(['age']).result.count()
df_female_age_count = pd.DataFrame(female_age_count).reset_index()
df_male_age_count = pd.DataFrame(male_age_count).reset_index()
df_female_count = pd.merge(age_female,df_female_age_count,on = 'age')
df_female_count['percent'] = round(df_female_count['result_x']/df_female_count['result_y'],2)
df_male_count = pd.merge(age_male,df_male_age_count,on = 'age')
df_male_count['percent'] = round(df_male_count['result_x']/df_male_count['result_y'],2)
df_male_count
```

```python
age_gender_line = (
    Line({"theme": ThemeType.ROMA})
    .add_xaxis(xaxis_data = df_female_count['age'].tolist())
    .add_yaxis(
        series_name = "女",
        y_axis = df_female_count['percent'].tolist()
    )
    .add_yaxis(
        series_name = "男",
        y_axis=df_male_count['percent'].tolist()
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="age_gender_Class"),
        tooltip_opts=opts.TooltipOpts(trigger="axis"),
        toolbox_opts=opts.ToolboxOpts(is_show=True),
        xaxis_opts=opts.AxisOpts(type_="category", boundary_gap=False),
        datazoom_opts=[opts.DataZoomOpts()],
    )
)
```

```python
age_gender_line.render_notebook()
```
